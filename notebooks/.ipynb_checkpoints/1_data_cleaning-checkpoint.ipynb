# Malaria Prediction in Nigerian LGAs

## Project Overview

This project aims to understand how malaria affects Local Government Areas (LGAs) 
across Nigeria and build a machine learning model to predict total malaria cases.

## Dataset

**Source**: [Africa Synth Malaria Dataset](https://huggingface.co/datasets/electricsheepafrica/africa-synth-malaria-malaria-burden-spatiotemporal-all)

**License**: CC BY 4.0

## Aims & Objectives

1. **Exploratory Analysis** — Understand patterns in malaria distribution
2. **Identify Key Drivers** — Find which factors drive cases (rainfall, temperature, season, etc.)
3. **Build Prediction Model** — Train ML model with R² > 0.75
4. **Feature Importance** — Understand which features the model relies on most

## Project Structure

1. **1_data_loading_and_cleaning.ipynb** — Load and clean data
2. **2_eda.ipynb** — Exploratory Data Analysis
3. **3_modelling_and_tuning.ipynb** — Train and evaluate models

## 1. Import Libraries

```python
#Importing Libraries

import numpy as np
import pandas as pd
```

## 2. Load Data

```python
#Loading the dataset

df = pd.read_csv("data_set/malaria_burden_moderate_seed.csv")
df = df.replace(r'^\s*$', np.nan, regex = True, inplace = True)
df.head()
```

## 3. Data Cleaning

Remove unnecessary columns and handle missing values.

```python
# Check initial data types and shape

print("Original Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
```

## 4. Feature Selection

Select only the columns needed for malaria prediction modelling. 
Drop case breakdowns (symptomatic, asymptomatic, age groups, etc.) 
and keep only the total cases target variable.

```python
# Drop unnecessary columns
columns_to_drop = [
    'record_id',
    'source', 
    'symptomatic_cases',	
    'asymptomatic_cases',	
    'cases_under5',	
    'cases_5to15',
    'cases_15plus',	
    'suspected_cases',	
    'tested_cases',	
    'positive_cases',
    'rd_t_positive',	
    'microscopy_positive',
    'pcr_positive',
    'predicted_cases_lower_ci',
    'predicted_cases_upper_ci'
]

df = df.drop(columns=columns_to_drop)

print(f"Shape after dropping unnecessary columns: {df.shape}")
print(f"\nRemaining columns:\n{df.columns.tolist()}")
```

## 5. Handle Missing Values

Check for missing values and remove rows with NaN entries.

```python
# Check missing values before cleaning

print("Missing values before cleaning:")
print(df.isnull().sum())
print(f"\nTotal rows with missing values: {df.isnull().any(axis=1).sum()}")
```

```python
# Remove rows with missing values
df = df.dropna()

# Verify no missing values remain
print("Missing values after dropna():")
print(df.isnull().sum())
```

## 6. Remove Duplicates

Remove any duplicate rows from the dataset.

```python
# Remove duplicate rows
print(f"Rows before removing duplicates: {df.shape[0]}")

df = df.drop_duplicates()

print(f"Rows after removing duplicates: {df.shape[0]}")
print(f"\nFinal cleaned dataset shape: {df.shape}")
```

## 7. Save Cleaned Data

Save the cleaned dataset for use in EDA and modelling.

```python
# Save cleaned dataset
df.to_csv('data_set/cleaned_malaria_burden_moderate_seed.csv', index=False)

print(f"Cleaned data saved to: 'data_set/cleaned_malaria_burden_moderate_seed.csv'")
print(f"Final dataset: {df.shape[0]} rows, {df.shape[1]} columns")
```
