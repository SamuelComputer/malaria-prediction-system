# Exploratory Data Analysis (EDA)

## Overview

This notebook performs exploratory data analysis on the malaria dataset to understand 
patterns in malaria distribution across Nigerian LGAs and identify key drivers of cases.


## Analysis Structure

1. **Distribution Analysis** — How are cases distributed?
2. **Geographic Analysis** — Which states and LGAs are most affected?
3. **Temporal Analysis** — How do cases vary by year, month, and season?
4. **Environmental Analysis** — How do rainfall and temperature affect cases?
5. **Intervention Analysis** — Does intervention reduce cases?
6. **Correlation Analysis** — Which features correlate most with cases?

```python
#Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Loading the dataset

df = pd.read_csv("data_set/cleaned_malaria_burden_moderate_seed.csv")

df.head()
```

## Question 1: 
Are malaria cases evenly spread or are there extreme hotspots?

```python
## Plot 1 — Distribution of Total Malaria Cases

plt.figure(figsize=(6, 4))
sns.histplot(df['total_cases'], bins=30, kde=True, color='orange')
plt.title('Distribution of Total Malaria Cases')
plt.xlabel('Total Cases')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Distribution of Total Malaria Cases

The distribution of total malaria cases is heavily right skewed. 
Most LGAs record relatively low case counts, while a small number 
of LGAs experience extremely high burdens — suggesting geographic 
hotspots that may require targeted intervention.

## Question 2: 
Which states have the highest malaria burden in Nigeria?

```python
top_states = df.groupby('state')['total_cases'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(6, 4))
sns.barplot(x=top_states.values, y=top_states.index, 
            hue=top_states.index, palette='Reds_r', legend=False)
plt.title('Top 10 States by Total Malaria Cases')
plt.xlabel('Total Cases')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

## Top 10 States by Total Malaria Cases

Southern states dominate the top 10, with Delta, Cross River, Benue, Imo, Bayelsa, Anambra and Enugu recording the highest malaria burden. This is likely driven by higher rainfall and dense vegetation in the south, which create ideal mosquito breeding conditions.

Notably, northern states like Borno, Jigawa and Bauchi also appear in the top 10, which may be linked to population size and reporting completeness rather than environmental factors one.

## Question 3: 
Which specific LGAs are the biggest malaria hotspots?

```python
top_lga = df.groupby('lga')['total_cases'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(6, 3))
sns.barplot(x=top_lga.values, y=top_lga.index, 
            hue=top_lga.index, palette='Reds_r', legend=False)
plt.title('Top 10 LGA by Total Malaria Cases')
plt.xlabel('Total Cases')
plt.ylabel('LGA')
plt.tight_layout()
plt.show()
```

## Top 10 LGAs by Total Malaria Cases

Delta state dominates the top 10, with Delta_LGA3, Delta_LGA2 and Delta_LGA1 recording the highest malaria burden, each exceeding 3 million cumulative cases. Cross River and Benue 
also have multiple LGAs in the top 10.

This confirms that malaria burden is not evenly spread across LGAs — a small number of LGAs, particularly in Delta and Cross River states, are responsible for a disproportionately high share of total cases and should be prioritized for intervention.

## Question 4: 
Is malaria getting better or worse over time?

```python
cases_by_year = df.groupby('year')['total_cases'].sum()

plt.figure(figsize=(6, 4))
sns.lineplot(x=cases_by_year.index, y=cases_by_year.values, marker='o', color='red')
plt.title('Total Malaria Cases by Year')
plt.xlabel('Year')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Total Malaria Cases by Year

Malaria cases initially rose before experiencing a brief decline, suggesting possible improvements in healthcare intervention during that period. However, cases spiked sharply in 2024, the highest  recorded in the dataset, which may be linked to increased rainfall, reduced intervention coverage, or improved case reporting capturing more cases than before.

The declining trend after 2024 is encouraging and may indicate that interventions were scaled up in response to the spike.

## Question 5: 
Which months have the highest malaria burden?

```python
cases_by_month = df.groupby('month')['total_cases'].sum()

plt.figure(figsize=(6, 4))
sns.barplot(x=cases_by_month.index, y=cases_by_month.values, color='red')
plt.title('Total Malaria Cases by Month')
plt.xlabel('Month')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Total Malaria Cases by Month

Malaria cases are lowest in the first three months (January-February) and last two months (November-December), corresponding to Nigeria's dry season when mosquito breeding is limited.

Cases rise sharply from April and peak in months 4 (April) and 9 (September) — the height of the rainy season. Waterlogged environments during this period create ideal breeding grounds for mosquitoes, driving the surge in malaria cases.

This strong seasonal pattern confirms that rainfall is a key driver of malaria burden in Nigerian LGAs.

## Question 6: 
How much does season affect malaria cases?

```python
cases_by_season= df.groupby('season')['total_cases'].sum()

plt.figure(figsize=(6, 4))
sns.barplot(x=cases_by_season.index, y=cases_by_season.values, color='red')
plt.title('Total Malaria Cases by season')
plt.xlabel('season')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Total Malaria Cases by Season

The rainy season records significantly higher malaria cases compared to the dry season, confirming the seasonal pattern observed in the monthly analysis. Waterlogged environments during the rainy season create ideal mosquito breeding conditions, driving the surge in cases.

## Question 7: 
Does more rainfall lead to more malaria cases?

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['rainfall_mm'], y=df['total_cases'], alpha=0.4, color='red')
plt.title('Rainfall vs Total Malaria Cases')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Rainfall vs Total Malaria Cases

The scatter plot reveals a clear positive relationship between rainfall and total malaria cases. LGAs with low rainfall (0-200mm) consistently record low case counts, while LGAs with high rainfall (300-600mm) show a dramatic spread in cases — with some exceeding 175,000.

This confirms rainfall as one of the strongest drivers of malaria burden in Nigerian LGAs, supporting the seasonal patterns observed in earlier plots.

## Question 8: 
How does temperature relate to malaria cases?

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['temperature_avg_c'], y=df['total_cases'], alpha=0.4, color='orange')
plt.title('Temperature vs Total Malaria Cases')
plt.xlabel('Temperature (°C)')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Temperature vs Total Malaria Cases

Similar to rainfall, temperature shows a positive relationship with total malaria cases. LGAs with lower temperatures (23-26°C) record consistently low cases, while warmer temperatures (28-32°C) are associated with significantly higher case counts.

This is biologically consistent, mosquitoes are cold blooded and breed more actively in warmer conditions, accelerating malaria transmission in higher temperature LGAs.

## Question 9: 
Does intervention reduce malaria cases?

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['intervention_flag'], y=df['total_cases'], 
            hue=df['intervention_flag'], palette='Reds', legend=False)
plt.title('Intervention Flag vs Total Malaria Cases')
plt.xlabel('Intervention Flag (0 = No, 1 = Yes)')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

## Intervention Flag vs Total Malaria Cases

LGAs with no intervention (0) show significantly higher malaria cases and more extreme outliers — with some exceeding 175,000 cases. In contrast, LGAs with intervention (1) record lower median cases and far fewer extreme values, with a maximum around 65,000.

This suggests that malaria interventions are effective in reducing both the overall burden and the occurrence of extreme hotspots in Nigerian LGAs.

## Question 10: 
Which numerical features are most correlated with total malaria cases?

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), 
            annot=True, 
            fmt='.2f', 
            cmap='Reds',
            linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

The heatmap confirms rainfall_mm (0.57) as the strongest predictor of total malaria cases, followed by vegetation_index (0.44) and reporting_completeness_pct (0.39). Notably, intervention_flag shows a negative correlation (-0.11), confirming that interventions reduce malaria burden. Year and month show almost no direct linear correlation with cases.
